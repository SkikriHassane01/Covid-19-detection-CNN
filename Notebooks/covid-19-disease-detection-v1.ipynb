```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import math
import os

import pickle

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# deep learning
import tensorflow as tf
from tensorflow.keras.layers import GlobalAveragePooling2D, Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.vgg16 import VGG16

from tensorflow.keras.layers import RandomRotation, RandomContrast, RandomZoom, RandomFlip, RandomTranslation

from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.optimizers import Adam

# Training Model callbacks
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler, ModelCheckpoint

%matplotlib inline
plt.style.use('ggplot')
```

```python
strategy = tf.distribute.MirroredStrategy()
BATCH_SIZE = 16 * strategy.num_replicas_in_sync
height, width = 224, 224
```

```python
path_train = "/kaggle/input/covid19-image-dataset/Covid19-dataset/train"
path_test = "/kaggle/input/covid19-image-dataset/Covid19-dataset/test"

def generate_data(DIR):
    datagen = ImageDataGenerator(rescale=1./255.)
    
    generator = datagen.flow_from_directory(
        DIR,
        batch_size=BATCH_SIZE,
        shuffle=True,
        seed=42,
        class_mode='binary',
        target_size=(height,width),
        classes={'Normal': 0, 'Viral Pneumonia': 1,'Covid': 2}
    )
    return generator


train_ds = generate_data(path_train)
test_ds = generate_data(path_test)

total_image = np.concatenate([train_ds.labels,test_ds.labels])

print('\n\n',{'Normal_cases':len(np.where(total_image==0)[0]),
      'Viral_Pneumonia_cases':len(np.where(total_image==1)[0]),
             'Covid_cases':len(np.where(total_image==2)[0])})
```

```python
def plot_pie_chart(distribution, title, ax):
    distribution_series = pd.Series(distribution)
    counts = distribution_series.value_counts()
    ax.pie(counts, labels=[f'{label} : {count}' for label, count in zip(counts.index, counts)],
           autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
    ax.axis('equal')  
    ax.set_title(title,pad=20)

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

plot_pie_chart(train_ds.labels, "Training Data Distribution", axes[0])
plot_pie_chart(test_ds.labels, "Testing Data Distribution", axes[1])

plt.tight_layout()
plt.show()
```

```python
def image_plot(generator,image_numbers):
    img_feature = generator[0][0][:image_numbers]
    img_label = generator[0][1][:image_numbers]

    plt.figure(figsize=(20, 15))
    for i in range(image_numbers):
        ax = plt.subplot(math.ceil(image_numbers/4),4, i + 1)
        plt.imshow(img_feature[i])
        plt.title("Normal" if img_label[i] == 0 else "Viral Pneumonia" if img_label[i] == 1 else "Covid")
        plt.axis("off")

image_plot(train_ds,15)
```

    <hr>
    <p>
<code>What is Data Augmentation</code>
<p>is a technique used to prevent overfitting and handling the imbalance in a dataset in machine learning by artificially increasing the size of a training dataset. This involves applying small modifications to the existing data, which helps the model learn to recognize variations in the data while keeping the labels the same. Common transformation methods include converting images to grayscale, flipping them horizontally or vertically, cropping them randomly, adjusting colors, and rotating them. By using these methods, you can significantly increase the number of training examples, making your model more robust and improving its performance.</p>
</div>


```python
def generate_data_augmented(DIR):
    datagen = ImageDataGenerator(
        rescale=1./255.,
        zoom_range=0.1,
        rotation_range=20,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip = True
    )
    generator = datagen.flow_from_directory(
        path_train,
        batch_size=BATCH_SIZE,
        seed=42,
        class_mode='binary',
        target_size=(height, width),
        classes={'Normal': 0, 'Viral Pneumonia': 1,'Covid': 2}
    )
    return generator

aug_train_ds = generate_data_augmented(path_train)
```

    🔻  Building CNN Model 🚊🔻</h1>

```python
tf.keras.backend.clear_session()

input_shape = (height, width, 3)



base_model = VGG16(
    weights='imagenet', 
    include_top=False,
    input_shape=input_shape
)

base_model.trainable = False

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(3, activation='softmax')
])


model.compile(
    loss='sparse_categorical_crossentropy', 
    optimizer= Adam(0.001),
    metrics=['acc']
)
```

```python
# Callbacks 
    
def exponential_decay(lr0, s):
    def exponential_decay_fn(epoch):
        return lr0 * 0.1 **(epoch / s)
    return exponential_decay_fn

exponential_decay_fn = exponential_decay(0.01, 20)

lr_scheduler = LearningRateScheduler(exponential_decay_fn)

checkpoint = ModelCheckpoint("Covid19_model.keras", monitor='acc', verbose=1, mode='max',save_best_only=True)

early_stopping =EarlyStopping(monitor="val_loss", mode="min",restore_best_weights=True, patience=5)
```

```python
history = model.fit(
    train_ds,
    validation_data=test_ds,
    shuffle=False, 
    verbose=True,
    callbacks=[checkpoint, early_stopping, lr_scheduler],
    epochs=100
)
```

```python
train_result = model.evaluate(train_ds)
test_result = model.evaluate(test_ds)

no_augmented_df = pd.DataFrame(zip(train_result,test_result),columns=['Train','Val'],index=['Loss','Acc'])
no_augmented_df
```

```python
# Save the history
with open('training_history.pkl', 'wb') as file:
    pickle.dump(history.history, file)
    
    
# Load the training history
with open('training_history.pkl', 'rb') as file:
    history = pickle.load(file)
```

```python
# Extract the data
acc = history['acc']
val_acc = history['val_acc']
loss = history['loss']
val_loss = history['val_loss']
epochs = range(len(acc))

# Plot accuracy
fig = plt.figure(figsize=(13,7))
plt.title("training vs testing acuuracy")
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot loss
fig = plt.figure(figsize=(13,7))
plt.title("training vs testing loss")
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.legend(loc='upper left')
plt.show()
```

```python
import math
ytest = np.array([])
xtest = []

for i in range(math.ceil(len(test_ds.classes)/BATCH_SIZE)):
    xtest.append(test_ds[i][0]) 
    ytest= np.concatenate((ytest,test_ds[i][-1])) 
    
xtest = np.concatenate((xtest),axis=0)

ypred_prob =model.predict(xtest)
ypred = np.argmax(ypred_prob,axis=1)

plt.figure(figsize=(6, 6))
hm = sns.heatmap(confusion_matrix(ytest,ypred), annot=True, vmin=0, fmt='g', cmap='Blues', cbar=False,
            xticklabels=['Normal','Viral Penumonia','Covid'],yticklabels=['Normal','Viral Penumonia','Covid'])  
hm.set(xlabel='Predicted_labels')
hm.set(ylabel='True_labels')
print(classification_report(ytest,ypred))
```
